## Label Encoding

- 범주형 데이터를 숫자형 데이터로 변환

- 각 범주에 대해 고유한 숫자 부여

```python
import pandas as pd

data = pd.DataFrame({
    'Team': ['삼성', 'LG', '롯데', '삼성', '두산', '롯데', 'LG', '두산', '삼성']
})

print(data)
```

### 1. Pandas로 수동 구현

```python
mapping = {'삼성': 2, 'LG': 1, '롯데': 0, '두산': 3}
data['Team_encoded_manual'] = data['Team'].map(mapping)

print(data)
```

### 2. Scikit-Learn


```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

data['Team_encoded'] = encoder.fit_transform(data['Team'])

print(data)
```
